```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "https://api.openalex.org/works?filter=institutions.country_code:{},institutions.country_code:{}"

# List of values for "country_code_1"
country_codes_1 = ["ru"]  # Source country code

# List of values for "country_code_2"
country_codes_2 = ["de", "pl", "ua"]  # Target country code

# List to store the retrieved metadata counts
metadata_counts = []

# Iterate through the lists of "country_code_1" and "country_code_2"
for code_1 in country_codes_1:
    for code_2 in country_codes_2:
        # Create the API URL with the variables
        api_url = base_url.format(code_1, code_2)

        # Make the API request
        response = requests.get(api_url)

        # Check if the response is successful (status code 200)
        if response.status_code == 200:
            data = response.json()
            
            # Get the count from the response
            count = data["meta"]["count"]
            
            # Append the country and count to the list
            metadata_counts.append({"Country": f"{code_1} to {code_2}", "Count": count})

# Create a DataFrame from the metadata counts
metadata_df = pd.DataFrame(metadata_counts)

# Print the resulting DataFrame
print(metadata_df)
```

```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "https://api.openalex.org/works?filter=institutions.country_code:{},publication_year:"

# List of values for "country_code_1"
country_codes_1 = ["ru"]  # Replace with your specific values

# List of values for "country_code_2"
country_codes_2 = ["de", "pl", "ua", "in", "nl", "us"]  # Replace with your specific values

# List of years to iterate through
years = list(range(2017, 2024))

# List to store the retrieved metadata counts
metadata_counts = []

# Iterate through the lists of "country_code_1", "country_code_2", and years
for code_1 in country_codes_1:
    for code_2 in country_codes_2:
        for year in years:
            # Create the API URL with the variables
            api_url = base_url.format(code_1) + str(year)

            # Make the API request
            response = requests.get(api_url)

            # Check if the response is successful (status code 200)
            if response.status_code == 200:
                data = response.json()
                
                # Get the count from the response
                count = data["meta"]["count"]
                
                # Append the country, year, and count to the list
                metadata_counts.append({"Country": f"{code_1} to {code_2}", "Year": year, "Count": count})

# Create a DataFrame from the metadata counts
metadata_df = pd.DataFrame(metadata_counts)

# Print the resulting DataFrame
print(metadata_df)
```
